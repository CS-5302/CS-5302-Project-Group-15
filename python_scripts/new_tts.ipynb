```python
from transformers import BarkModel, BarkProcessor
from langdetect import detect
from IPython.display import Audio
```

```python
# Initialize model and processor
model_name = "suno/bark-small"
model = BarkModel.from_pretrained(model_name)
processor = BarkProcessor.from_pretrained(model_name)
sampling_rate = model.generation_config.sample_rate
```

```python

language_to_voice_preset = {
    "en": "v2/en_speaker_3",  # English
    "fr": "v2/fr_speaker_1",  # French
    
}
```

```python
def text_to_speech(text):
   
    detected_lang = detect(text)
    print(f"Detected language: {detected_lang}")
    voice_preset = language_to_voice_preset.get(detected_lang, "v2/en_speaker_3")  
    inputs = processor(text, voice_preset=voice_preset)
    speech_output = model.generate(**inputs).cpu().numpy()
    return Audio(speech_output[0], rate=sampling_rate)

```

```python
# Example usage:
text_en = "I am Maryam!"
text_to_speech(text_en)
```
