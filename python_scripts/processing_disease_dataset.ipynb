```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Disease_symptom_and_patient_profile_dataset.csv")
df
```

```python
df.drop(["Age","Gender"],axis=1,inplace=True)
df
```

```python
prompts = []
output = []
symptoms = []

for index,row in df.iterrows():
    symptoms.append([col for col, value in row.items() if value == "Yes" or value=="High"])
    prompts.append("I have " + ", ".join(symptoms[index]) + ". What disease do i have?")
    
    if row["Outcome Variable"] == "Positive":
        result = f"You have {row['Disease']}."
    else:
        result = "You don't have a disease."

    output.append(result)

df['prompts'] = prompts
df['results'] = output

df
```
